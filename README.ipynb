# Versions of libraries programs used. 

# Description of Contents

create_data.py allows users to create new input data to test the two algorithms. 

type:  
    
    python create_data.py ___ ___ 
    
    to specify mean for poisson distribution and specify the probabilities of each drink being ordered. 

fifo.py runs the fifo algorithm solution on specified input file

type: 

    python fifo.py ____ 
    
    to run the algorithm on input file

type: 

    python fifo.py test 
    
    to run tests on the fifo.py algorithm
    
optimized.py runs the optimized algorithm solution on specified input file

type: 
    
    python optimized.py ___
    
    to run the algorithm on input file
    
type: 

    python optimized.py tests 
    
    to run tests on optimzied.py
    
    
metric.py displays metrics for specific algorithm. It will be called by fifo.py and optimized.py 

type: 
    
    python metric.py test
    
    to run tests on metric.py
    
fifo_tests is a folder containing all of the test files for fifo.py

input_files is a folder containing all of the input files that can be run by either algorithm

metric_tests is a folder containing all of the test files for metric.py

optimized_tests is a folder containing all of the test files for optimized.py

output_files is a folder containing all of the output files generated by fifo.py, optimized.py, and metric.py (both fifo and optimized algorithm generates files with *_metric_output.json (where * is the name of the algorithm), which is simply a file to help metric.py calculate metrics)
